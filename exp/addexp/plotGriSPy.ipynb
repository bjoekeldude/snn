```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('bmh')
```

```python
with open('result/add1_snn_index_time.npy', 'rb') as f:
    snn_index_time1 = np.load(f)
    
with open('result/add1_snn_query_time.npy', 'rb') as f:
    snn_query_time1 = np.load(f)
    
with open('result/add1_gp_index_time.npy', 'rb') as f:
    gp_index_time1 = np.load(f)

with open('result/add1_gp_query_time.npy', 'rb') as f:
    gp_query_time1 = np.load(f)
```

```python
with open('result/add2_snn_index_time.npy', 'rb') as f:
    snn_index_time2 = np.load(f)
    
with open('result/add2_snn_query_time.npy', 'rb') as f:
    snn_query_time2 = np.load(f)
    
with open('result/add2_gp_index_time.npy', 'rb') as f:
    gp_index_time2 = np.load(f)

with open('result/add2_gp_query_time.npy', 'rb') as f:
    gp_query_time2 = np.load(f)
```

```python
x_range = [2, 3, 4]
plt.figure(figsize=(7, 5.2))
plt.rcParams['axes.facecolor'] = 'white'
plt.loglog(x_range,
         gp_index_time1.mean(axis=1), marker="p", markerfacecolor='yellowgreen',
         markersize=12, linestyle=':', c='k', linewidth=2, label='GriSPy', )
plt.loglog(x_range,
         snn_index_time1.mean(axis=1), marker="s", markerfacecolor='none', 
         markersize=10, linestyle='--', c='chocolate', linewidth=2,label='SNN')
plt.legend(fontsize=15, ncol=3, bbox_to_anchor=(0., 0, 0.75, -0.16))
plt.title("index time", fontsize=20)
plt.xlabel("dimension", fontsize=20)
plt.grid(True)
# plt.ylim(0.001, max(index_time_kd)+0.05)
plt.tick_params(axis='both',  labelsize=15)
plt.xticks(x_range) 

plt.savefig('result/indexCompGriSPy.pdf', bbox_inches='tight')
plt.show()
```

```python
x_range = [2, 3, 4]
plt.figure(figsize=(7, 5.2))
plt.rcParams['axes.facecolor'] = 'white'
plt.loglog(x_range,
         gp_query_time1.mean(axis=1), marker="p", markerfacecolor='yellowgreen',
         markersize=12, linestyle=':', c='k', linewidth=2, label='GriSPy', )
plt.loglog(x_range,
         snn_query_time1.mean(axis=1), marker="s", markerfacecolor='none', 
         markersize=10, linestyle='--', c='chocolate', linewidth=2,label='SNN')
plt.legend(fontsize=15, ncol=3, bbox_to_anchor=(0., 0, 0.75, -0.16))
plt.title("query time", fontsize=20)
plt.xlabel("dimension", fontsize=20)
plt.grid(True)
# plt.ylim(0.001, max(index_time_kd)+0.05)
plt.tick_params(axis='both',  labelsize=15)
plt.xticks(x_range) 

plt.savefig('result/queryCompGriSPy.pdf', bbox_inches='tight')
plt.show()
```
